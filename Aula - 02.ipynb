```python

lista = [1, 2, 4, 5]

def imprimir_primeiro(lista):
    print(lista[0])

imprimir_primeiro(lista)

#Complexidade o(1) porque o acesso ao primeiro elemento da lista é constante, independente do tamanho da lista
```

```python

def encontrar_maximo(lista):
    valor_maximo = lista[0]
    for numero in lista:
        if numero > valor_maximo:
            valor_maximo = numero
    return valor_maximo

encontrar_maximo(lista)

#Complexidade o(n) porque tem que percorrer toda a lista para encontrar o máximo
```

```python


def busca_binaria(lista, alvo):
    esquerda = 0
    direita = len(lista) - 1
    while esquerda <= direita:
        meio = (esquerda + direita) // 2
        if lista[meio] == alvo:
            return meio
        elif lista[meio] < alvo:
            esquerda = meio + 1
        else:
            direita = meio - 1
    return -1


busca_binaria(lista, 4)

# O(n log n) por causa da ordenação da lista, a busca binária é o(log n) mas a ordenação é o(n log n)
```

```python
def fib(n):
    if n < 2:
        return n
    else:
        return fib(n-1) + fib(n-2)

fib(2)

# o (2*n) porque cada chamada de fib gera duas chamadas, uma para n-1 e outra para n-2, e isso se repete até chegar em n < 2, onde retorna n. Portanto, o número de chamadas cresce exponencialmente com n.
```

```python
#projeto 2 - ver se ta certo - A complexidade do algoritmo de recomendação de disciplinas é o(n*m) onde n é o número de alunos no histórico e m é o número de disciplinas em cada histórico, porque para cada aluno no histórico, o algoritmo precisa comparar as disciplinas com o histórico do aluno principal e isso pode levar até m comparações para cada um dos n alunos.
def recomendar_disciplinas(historico_alunos, historico_aluno_principal):
    recomendacoes = set()
    for historico_outro_aluno in historico_alunos:
        intersecao = set(historico_aluno_principal) & set(historico_outro_aluno)
        if intersecao:
            diferenca = set(historico_outro_aluno) - set(historico_aluno_principal)
            recomendacoes.update(diferenca)
    return list(recomendacoes)
```
