```python


def buscar_cliente(nome, lista_clientes):
    for i in range(len(lista_clientes)):
        if lista_clientes[i] == nome:
            return i
    return -1

clientes = ["Eduardo", "Ilda", "André", "Denis", "Ymyly"]
busca1 = "André"
busca2 = "Fafadebelem"
pos1 = buscar_cliente(busca1, clientes)
if pos1 != -1:
    print(f"cliente {busca1} encontrado na posicao {pos1}.")
pos2 = buscar_cliente(busca2, clientes)
if pos2 != -1:
    print(f"cliente {busca2} encontrado na posicao {pos2}.")
else:
    print(f"cliente {busca2} nao encontrado na lista")
```

```python
#5 
def conta_ocorrencias(valor, lista):
    contador = 0
    for i in lista:
        if i == valor:
            contador += 1
    return contador

print(conta_ocorrencias(3,[1,3,5,3,7]))
print(conta_ocorrencias("a", ["a", "b", "c", "a"]))
```

```python
 
def busca_sequencial(catalogo, titulo):
    for i in range(len(catalogo)):
        if catalogo[i] == titulo:
            return i
    return -1


def buscar_filme(catalogo, titulo):
    indice = busca_sequencial(catalogo, titulo)
    if indice != -1:
        return f"filme {titulo} encontrado na posicao {indice} do catalogo"
    else:
        return f"filme {titulo} nao encontrado"

catalogo_filmes = ["Frozen", "Zootopia", "Elite2" ]
titulo_buscado = input("Digite o titulo do filme que voce procura:")
resultado_busca = buscar_filme(catalogo_filmes, titulo_buscado)
print(resultado_busca)
```
